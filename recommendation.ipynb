```python
f=open("stopwords.txt",'r')
stopwords=f.read().splitlines()
```

```python
# few words which are having high in frequency and low in impact are called stopwords including is,are etc.
# so we download that from nltk library and after remove that from our text
# there may be many long word that meaning is same as root word. The conversion of long word to its root word called stemmering
stemmer = nltk.SnowballStemmer("english")
```

```python
# clean the data
import re
def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
```

```python
required_df["Title"]=required_df["Title"].apply(clean)
```

```python
print(required_df["Title"])
```

```python
genreList=required_df["Genres"].to_list()
# making a tfidf(term frequency and inverse document frequency) matrix
# each row represent a document
# each column represent different terms which is associated with that document
# and value represent how closely they are related
tfidf=text.TfidfVectorizer(stop_words="english")
tfidf_matrix = tfidf.fit_transform(genreList)
similarity = cosine_similarity(tfidf_matrix)
# we find the angle with the help of cosine similarity if toward 0 means high correlated
```

```python
indices=pd.Series(required_df.index,index=required_df["Title"]).drop_duplicates()
```

```python
def recommendation(title,similarity=similarity):
    index=indices[title]
    similarity_score=list(enumerate(similarity[index]))
    similarity_score = sorted(similarity_score, key=lambda x: x[1], reverse=True)
    similarity_score = similarity_score[0:10]
    movieindices = [i[0] for i in similarity_score]
    return required_df['Title'].iloc[movieindices]
```
